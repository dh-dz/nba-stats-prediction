```python
import pandas as pd
import numpy as np

# Load CSVs
#odds_df = pd.read_csv("odds_select.csv")
merged_df = pd.read_csv("merged_data.csv")
#teams_df = pd.read_csv("teams.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define stat columns to compute home - away differences
stats = ['PTS', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'AST', 'REB']

# Compute and plot histograms for each stat difference
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for i, stat in enumerate(stats):
    diff_col = f'{stat}_diff'
    merged_df[diff_col] = merged_df[f'{stat}_home'] - merged_df[f'{stat}_away']
    sns.histplot(merged_df[diff_col].dropna(), kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {stat} Difference (Home - Away)')
    axes[i].set_xlabel(f'{stat} Difference')

plt.tight_layout()
plt.show()
```

```python
# Copy merged_df into eda_df for EDA-specific analysis
eda_df = merged_df.copy()

# Construct home and away moneylines based on teamA_home
eda_df['home_moneyLine'] = np.where(eda_df['teamA_home'] == 1,
                                    eda_df['teamA_moneyLine'],
                                    eda_df['teamB_moneyLine'])

eda_df['away_moneyLine'] = np.where(eda_df['teamA_home'] == 1,
                                    eda_df['teamB_moneyLine'],
                                    eda_df['teamA_moneyLine'])

# Compute home - away moneyline difference
eda_df['moneyline_diff_home_away'] = eda_df['home_moneyLine'] - eda_df['away_moneyLine']

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.histplot(eda_df['moneyline_diff_home_away'].dropna(), kde=True, bins=50)
plt.title("Distribution of Moneyline Difference (Home - Away)")
plt.xlabel("Moneyline Difference (Home - Away)")
plt.ylabel("Frequency")
#plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of score difference vs. whether team A is the home team
plt.figure(figsize=(8, 6))
sns.stripplot(data=eda_df, x='teamA_home', y='score_diff', jitter=True, alpha=0.5,
              palette={'0': 'steelblue', '1': 'darkorange'}, marker='o')

plt.title("Score Difference (A - B) vs. Team A Home Indicator")
plt.xlabel("Team A is Home (0 = Away, 1 = Home)")
plt.ylabel("Score Difference (A - B)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
